```python
df1 = pd.DataFrame(table1)
df2 = pd.DataFrame(table2)
columns1 = ['student_name','maths','physics']
columns2 = ['student_name','chemistry','english']

df1.columns = columns1
df2.columns = columns2

df1['student_name'] = ['a','b','c','d','e'] ## common elements a,c,e ## diff elements b,d
df2['student_name'] = ['x','y','c','a','e'] ## common elements a,c,e ## diff elements x,y

df1
df2
```

### Inner join
Take common elements b/w the two tables

```python
pd.merge(df1,df2,on=['student_name'],how='inner')
```

### Left join
Take everything on the left table and matching one's on the right. 

indicator parameter will show which rows are available in both and which one's are available only in the left table

```python
pd.merge(df1,df2,on=['student_name'],how='left')
```

```python
pd.merge(df1,df2,on=['student_name'],how='left',indicator=True) ## left outer join
```

### Right join
Take everything on the right table and matching one's on the left. 

indicator parameter will show which rows are available in both and which one's are available only in the right table

```python
pd.merge(df1,df2,on=['student_name'],how='right')
```

```python
pd.merge(df1,df2,on=['student_name'],how='right',indicator=True)
```

### Outer join
Takes all element from left table and all elements in the right table



```python
df3 = pd.merge(df1,df2,on=['student_name'],how='outer')
df3
```

```python
df3 = pd.merge(df1,df2,on=['student_name'],how='outer',indicator=True)
df3
```

```python
for col in df3.columns:
    df3.fillna(0,inplace=True)

df3
```

### Group By
Any groupby operation involves one of the following operations on the original object. They are,

- Splitting the Object
- Applying a function
- Combining the results

In many situations, we split the data into sets and we apply some functionality on each subset. In the apply functionality, we can perform the following operations,

- Aggregation − computing a summary statistic
- Transformation − perform some group-specific operation
- Filtration − discarding the data with some condition

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
   'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
df
```

```python
df.groupby('Team').groups
```

```python
df.groupby(['Team','Year']).groups
```

```python
grouped = df.groupby('Year')

for name,group in grouped:
    print('-> ',name)
    print(group)
```

```python
grouped.get_group(2014)
```

#### Aggregations
An aggregated function returns a single aggregated value for each group. Once the group by object is created, several aggregation operations can be performed on the grouped data.

An obvious one is aggregation via the aggregate or equivalent agg method

```python
grouped['Points'].agg(np.mean)
```

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
   'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)

grouped = df.groupby('Team')
grouped.agg(np.size)
```

```python
grouped['Points'].agg([np.sum, np.mean, np.std])
```

#### Transformations
Transformation on a group or a column returns an object that is indexed the same size of that is being grouped. Thus, the transform should return a result that is the same size as that of a group chunk.

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
   'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)

grouped = df.groupby('Team')
score = lambda x: (x - x.mean()) / x.std()*10
grouped.transform(score)
```

#### Filtration
Filtration filters the data on a defined criteria and returns the subset of data. The filter() function is used to filter the data.

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
   'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)

df.groupby('Team').filter(lambda x: len(x) >= 3)
```

## Stores dataset (Practice Problem)

```python
## customer id - 1 - 10000
## gender - 0/1
## age - 10 - 100
## store - 1 - 15
## Units - 10 - 50 (number of purchase items a month)
## Txn - 0 - 5 (number of times you visit the store)
## sale - 100 - 200 (Total value of transaction per month)
## basket value -- sales/txn
## basket size -- units/txn
## gross profit -- sales * 30%
## new/existing cust -- 0/1
```

```python
data = np.random.randint(0,1,(10000,11))
data.shape
```

```python
df = pd.DataFrame(data)
df.head()
df.columns = ['customer_id','gender','age','store','units','transaction','sale','basket_value','basket_size','gross_profit','new_exis']
#df['customer_id'] = df.index
out = ['cust_'+str(i) for i in range(0,10000)]
df['customer_id'] = out
df['gender'] = np.random.randint(0,2,10000)
df['age'] = np.random.randint(10,100,10000)
df['store'] = np.random.randint(1,16,10000)
df['units'] = np.random.randint(10,51,10000)
df['transaction'] = np.random.randint(1,6,10000)
df['sale'] = np.random.randint(100,201,10000)
df['basket_value'] = df['sale']/ df['transaction']
df['basket_size'] = df['units']/ df['transaction']
df['gross_profit'] = df['sale'] * .3
df['new_exis'] = np.random.randint(0,2,10000)

df

df.describe()
```

```python
df.groupby('store').sum()
df.groupby('store').mean()
df.groupby('store').count()
df.groupby('store').std()
```

```python
df.groupby('store').describe()['sale']
```

## Titanic dataset (Practice Problem)

```python
import seaborn as sns
df = sns.load_dataset('titanic')
df.head()
```

```python
class_fields = [i for i in df.columns if df[i].nunique() < 8]
class_fields
```

```python
for cf in class_fields:
    df.groupby(cf).mean()[['age','fare']]
    df.groupby(cf).std()[['age','fare']]
    df.groupby(cf).count()[['age','fare']]
df.to_excel()
```
