### Questions
- How does the price vary based on drive-wheel ?
- How does the price vary based on body-type ?
- How does the price vary based on width,height,length ?
- How is mileage and price related ?

### Observations
- For drive-wheel RWD price is higher
- For fuel-type diesel price is higher
- For body type hardtop and convertible price is higher
- Higher the engine-size higher the price
- For higher priced cars, less mileage on city and highway
- As the length and width of the car increases price increases
- No relationship between height, wheel base, stroke, horsepower

#### Positive correlation
- engine-size, length, width

#### Negative correlation
- City, Highway mileage

#### No Correlation
- Height, wheel base, stroke, horsepower
