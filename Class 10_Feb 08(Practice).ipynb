```python
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
table1 = np.random.randint(10,100,(5,3))
table2 = np.random.randint(10,100,(5,3))

df1 = pd.DataFrame(table1)
df2 = pd.DataFrame(table2)

columns1 = ['student_name','maths','physics']
columns2 = ['student_name','chemistry','biology']

df1.columns = columns1
df2.columns = columns2

df1['student_name'] = ['a','b','c','d','e']
df2['student_name'] = ['a','b','c','x','y']

df1
df2
```

```python
pd.merge(df1,df2,on=['student_name'],how='inner')
```

```python
pd.merge(df1,df2,on=['student_name'],how='left')
```

```python
pd.merge(df1,df2,on=['student_name'],how='right')
```

```python
pd.merge(df1,df2,on=['student_name'],how='outer')
```

```python
df3 = pd.merge(df1,df2,on=['student_name'],how='outer')

for col in df3.columns:
    df3.fillna(0,inplace=True)

df3
```

```python
pd.merge(df1,df2,on=['student_name'],how='left',indicator=True)
```

```python
pd.merge(df1,df2,on=['student_name'],how='right',indicator=True)
```

```python
pd.merge(df1,df2,on=['student_name'],how='outer',indicator=True)
```
