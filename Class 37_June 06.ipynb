```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
# create a black image
img = np.zeros((512,512,3), np.uint8)
# draw a diagonal blue line with thickness of 5 px
plt.subplot(2,2,1)

img = cv.line(img,(0,0),(511,511),(255,255,255),5)

plt.imshow(img)

plt.subplot(2,2,2)

img = cv.line(img,(0,0),(511,511),(255,0,0),5)

plt.imshow(img)
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
# create a white image
img = np.full((600, 600,3), 255)
#plt.subplot(2,2,1)
img = cv.line(img,(100,100),(300,300),(0,0,0),5)
plt.imshow(img)
```

```python
img = np.zeros((512,512,3),np.uint8)
cv.rectangle(img,(100,100),(400,500),(0,255,0),3)
plt.imshow(img)
```

```python
cv.circle(img,(447,63),63,(0,0,255),-1)
plt.imshow(img)
```

```python
cv.ellipse(img,(256,256),(100,50),0,0,360,255,-1)
plt.imshow(img)
```

```python
pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts = pts.reshape((-1,1,2))
cv.polylines(img,[pts],True,(0,255,255))
plt.imshow(img)
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
# Create a black image
img = np.ones((512,512,3), np.uint8)
img.fill(255)
# Draw a diagonal blue line with thickness of 5 px
img=cv.line(img,(100,100),(300,300),(0,0,0),10)
plt.imshow(img);
img=cv.rectangle(img, (100,100), (300,300), (255,0,0),10)
plt.imshow(img);
img=cv.rectangle(img, (50,50), (400,500), (0,255,0),10)
plt.imshow(img);
img=cv.circle(img, (440,60),60,  (0,0,255),10)
plt.imshow(img);
img=cv.ellipse(img,(120,100), (100,50) ,0,0,360, (255,0,255),10) 
plt.imshow(img);
```
